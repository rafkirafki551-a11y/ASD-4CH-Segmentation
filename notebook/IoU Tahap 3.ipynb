```python
import pandas as pd

# Path ke file Excel
file_path = "D:\\DATA RAFKI\\PunyaResti\\PunyaResti\\DATASET SMARTPHONE\\4CH\\TEST_4CH.xlsx"  # Ganti dengan path file kamu

# Membaca data dari Excel
df = pd.read_excel(file_path)

# Menghitung rata-rata IoU per objek
avg_iou_per_object = df.groupby("Object")["IoU"].mean().reset_index()

# Menampilkan hasil rata-rata per objek
print("\n=== Average IoU per Object ===")
for _, row in avg_iou_per_object.iterrows():
    print(f"{row['Object']}: Average IoU = {row['IoU']:.4f}")

# Menghitung overall average IoU
overall_avg_iou = df["IoU"].mean()

# Menampilkan overall average IoU
print("\n=== Overall Average IoU ===")
print(f"Overall Average IoU = {overall_avg_iou:.4f}")
```
