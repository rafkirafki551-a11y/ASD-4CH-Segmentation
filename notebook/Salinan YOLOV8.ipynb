#TRAINING DAN VALIDASI DATA

```python
from ultralytics import YOLO

# Load model
model = YOLO('yolov8l-seg.pt')  # Menggunakan model YOLOv8 n, s, m, l
```

```python
# Path ke data YAML
data_yaml = '/content/drive/MyDrive/4CH/4CH.yaml'  # Ganti dengan path ke file YAML data Anda
```

```python
# Parameter yang dapat disesuaikan
params = {
    'epochs': 50,          # Jumlah epoch
    'batch_size': 16,      # Ukuran batch
    'imgsz': 640,          # Ukuran gambar (default: 640x640)
    'optimizer': 'SGD',   # Optimizer: 'SGD', 'Adam', 'AdamW', dll.
    'learning_rate': 0.01, # Learning rate awal
}

# Latih model
model.train(
    data=data_yaml,
    epochs=params['epochs'],
    batch=params['batch_size'],
    imgsz=params['imgsz'],
    optimizer=params['optimizer'],
    lr0=params['learning_rate'],
    name='REV1_4CH_300_L_16',
    task='segment'
)
```

## UJI UNSEEN

```python
# Evaluasi data test
model.val(data="/content/drive/MyDrive/4CH/4CH.yaml", split='test')
```
